```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import netCDF4 as nc
from datetime import datetime, timedelta
from salishsea_tools import places

%matplotlib inline
```

```python
data = xr.open_dataset('/ocean/bmoorema/research/MEOPAR/analysis-ben/data/ONC/ubcONCTWDP1mV18-01_bd92_6265_f263.nc')
time = data.time.values.astype('datetime64[s]').astype(datetime)
```

```python
fig, axs = plt.subplots(3, 1, figsize=(17, 8))

xlim = [datetime(2017, 7, 5), datetime(2017, 7, 10)]

for ax, key in zip(axs, ['longitude', 'temperature', 'salinity']):
    ax.plot(time, data[key])
    ax.set_xlim(xlim)
```
