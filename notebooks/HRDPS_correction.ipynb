## HRDPS Correction Check
This notebook checks the HRDPS correction obtained from Luc Fillion's group

```python
from salishsea_tools import nc_tools, data_tools, visualisations, viz_tools
import xarray as xr
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
GEM_corr = xr.open_dataset('/results/forcing/atmospheric/GEM2.5/research/res_y2016m04d06.corr.nc')
GEM      = xr.open_dataset('/results/forcing/atmospheric/GEM2.5/research/res_y2016m04d06.nc')
```

```python
GEM_dict = {'lon': GEM.nav_lon, 'lat': GEM.nav_lat, 'u_wind': GEM.u_wind, 'v_wind': GEM.v_wind}
GEM_corr_dict = {'lon': GEM_corr.nav_lon, 'lat': GEM_corr.nav_lat, 'u_wind': GEM_corr.u_wind, 'v_wind': GEM_corr.v_wind}
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 10))

map_bounds=[-124, -122.7, 48.3, 49.7]

bathy = '/ocean/bmoorema/research/MEOPAR/NEMO-forcing/grid/bathy_meter_SalishSea2.nc'
    
# Plot land mask and coastline
viz_tools.plot_land_mask(ax, bathy, coords='map', color='burlywood')
viz_tools.plot_coastline(ax, bathy, coords='map')

Q = visualisations.plot_wind('2016 Apr 5 20:00', ax, 5, GEM_dict)
Q = visualisations.plot_wind('2016 Apr 5 20:00', ax, 5, GEM_corr_dict, color='red')

# Prettying
viz_tools.set_aspect(ax)
ax.set_xlim(map_bounds[0:2])
ax.set_ylim(map_bounds[2:4])
ax.grid()
```
